# Mathematics Behind Large Margin Classification

```python
%%latex
#https://www.coursera.org/learn/machine-learning/resources/Es9Qo
\begin{align}
& Vector \ Inner \ Product \\
& Say \ we\ have\ two\ vectors,\ u\ and\ v: \\
& u \ =\ \begin{bmatrix}\ u_1\ \newline\ u_2\ \end{bmatrix} \
 v \ =\ \begin{bmatrix}\ v_1\ \newline\ v_2\ \end{bmatrix}\\
& The\ length\ of\ vector\ v\ is\ denoted:||v||,\ and\ it\ describes\ the\ line\ on\ a\ graph\ from\ origin\ (0,0)\ to\ (v1,v2).\ \\
& The\ length\ of\ vector\ v\ can\ be\ calculated\ with\ \sqrt{v_1^2\ +\ v_2^2}\ the\ Pythagorean\ theorem.\ \\
& The\ projection\ of\ vector\ v\ onto\ vector\ u\ is\ found\ by\ taking\ a\ right\ angle\ from\ u\ to\ the\ end\ of\ v,\ creating\ a\ right\ triangle.\ \\

& 1) \ p\ =\ length\ of\ projection\ of\ v\ onto\ the\ vector\ u.\\
& 2) \ u^Tv=\ p\ \cdot\ ||u|| \\
& Note\ that\ u^Tv\ =\ ||u|| \cdot ||v|| \cos \theta\ where\ θ\ is\ the\ angle\ between\ u\ and\ v.\ Also,\ p=||v||cosθ.\\ 
& If\ you\ substitute\ p\ for\ ||v||cosθ,\ you\ get\ u^Tv= p \cdot ||u||.\\

& So\ the\ product\ u^Tv\ is\ equal\ to\ the\ length\ of\ the\ projection\ times\ the\ length\ of\ vector\ u. \\

& In\ our\ example,\ since\ u\ and\ v\ are\ vectors\ of\ the\ same\ length,\ u^Tv = v^Tu. \\

& u^Tv\ =\ v^Tu\ = p \cdot ||u||\ =\ u_1v_1\ +\ u_2v_2 \\
& If\ the\ angle\ between\ the\ lines\ for\ v\ and\ u\ is\ greater\ than\ 90\ degrees,\ then\ the\ projection\ p\ will\ be\ negative. \\
&\min_\Theta \dfrac{1}{2}\sum_{j=1}^n \Theta_j^2 \newline&= \dfrac{1}{2}(\Theta_1^2 + \Theta_2^2 + \dots + \Theta_n^2) \newline&= \dfrac{1}{2}(\sqrt{\Theta_1^2 + \Theta_2^2 + \dots + \Theta_n^2})^2 \newline&= \dfrac{1}{2}||\Theta ||^2 \newline
& We \ can \ use \ the \ same \ rules \ to \ rewrite \ \Theta^Tx^{(i)} \\

& \Theta^Tx^{(i)} \ = \ p^{(i)} \ \cdot \ ||\Theta \ || \ = \ \Theta_1x_1^{(i)} \ + \ \Theta_2x_2^{(i)} \ + \ \dots \ + \ \Theta_n \ x_n^{(i)} \\

& So \ we \ now \ have \ a \ new \ optimization \ objective \ by \ substituting \ p^{(i)} \ \cdot \ ||\Theta \ || \\


& p^{(i)} \ \cdot \ ||\Theta \ || \ \geq \ 1 \\
& p^{(i)} \ \cdot \ ||\Theta \ || \ \leq \ -1 \\

& The \ reason \ this \ causes \ a \ "large \ margin" \ is \ because: \ the \ vector \ for \ Θ \ is \ perpendicular \ to \ the \ decision \ boundary.\\
& In \ order \ for \ our \ optimization \ objective \ (above) \ to \ hold \ true, \ we \ need \ the \ absolute \ value \ of \ our \ projections \ p^{(i)} \ to \ be \ as \ large \ as \ possible. \\ 

& If \ Θ_0=0, \ then \ all \ our \ decision \ boundaries \ will \ intersect \ (0,0). \\ 
& If \ Θ_0≠0, \ the \ support \ vector \ machine \ will \ still \ find \ a \ large \ margin \ for \ the \ decision \ boundary. \\
\end{align}
```

## Logisit regression:

```python
%%latex
\begin{align}
& \it{J}(\theta)\ = \ min_{\theta} \ \frac{1}{\it{m}} \ \bigg[ \ \sum_{\it{i}=1}^{\it{m}} 
                                        \it{y}^{(i)} \big(-log(h_{\theta}(\it{x}^{(\it{i})})) \big)\ 
                                        + \ (1 \ - \ \it{y}^{(i)})
                                        \big(- \log(1 \ - \ h_{\theta}(\it{x}^{(\it{i})}) \big) \ \bigg] \ 
                                        + \ \frac{\lambda}{2\it{m}} \sum_{\it{j}=1}^{\it{n}}\theta_{\it{j}}^{2} \\
& \it{cost_{1}(\theta^{T}x)}\ = \ -log(h_{\theta}(\it{x}^{(\it{i})})) \\ 
& \it{cost_{0}(\theta^{T}x)}\ = \ -log(1 \ - \ h_{\theta}(\it{x}^{(\it{i})})) \\ 
& \bf{A + \lambda B} \ if \lambda\ is\ too\ high\ it\ may\ be\ high\ bias\ also\ as\ a\ low\ C: \\
& \bf{CA + B} \\
& After\ optimizing\ (remove\ \lambda-\ because\ it's\ a\ constant\ and\ add\ regularizing\ parametr\ C(C = \frac{1}{\lambda})\ we\ have\ formula:\\
& \it{J}(\theta)\ = \ C \sum_{\it{i}=1}^{\it{m}} \bigg[ \it{y}^{(i)}it{cost_{1}(\theta^{T}x)}\ + \ (1 \ - \ \it{y}^{(i)}) \ \it{cost_{0}(\theta^{T}x)} \bigg]
                      + \ \frac{1}{2} \sum_{\it{j}=1}^{\it{n}}\theta_{\it{j}}^{2} \\
\end{align}
```

```python
%matplotlib notebook

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

data = np.arange(-5,5,0.05)
hypotesis = 1 / (1 + np.exp(-data))

cost1 = -np.log(hypotesis)
cost0 = -np.log(1 - hypotesis)

c1 = np.maximum(np.zeros(len(data)), 0.6 * (1 - data))
c0 = np.maximum(np.zeros(len(data)), 0.6 * (1 + data))

plt.figure(1, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

plt.subplot(1,2,1)
plt.plot(data, cost1)
plt.plot(data, c1)
plt.grid('True')
plt.ylim(0,4)
plt.xlim(-3,3)

plt.subplot(1,2,2)
plt.plot(data, cost0)
plt.plot(data, c0)
plt.grid('True')
plt.ylim(0,4)
plt.xlim(-3,3)
```

# Logistic Regression vs. SVMsb

$\it{n}$ = number of features $(x \in \mathbb{R}^{n+1}$), $\it{m}$=number of training examples 

$\bf{1)}$ If $\it{n}$ is large(relative it $\it{m}$): (for example $\it{n}$ $\geqslant$ $\it{m}$, $\it{n}$=10000 and $\it{m}$ = 10...1000)

Then use logisitc regression, or $\bf{SVM}$ without a kernel("linear kernel")


$\bf{2)}$ If $\it{n}$ is small, and $\it{m}$ is intermediate: Use SVM with Gaussian kernel ($\it{n}$=1-1000, $\it{m}$=10-10000)

$\bf{3)}$ If $\it{n}$ is small, and $\it{m}$ is large: Create/add more features, then use logistic regression or SVM without a kernel ($\it{n}$=1-1000, $\it{m}$=50000+)

And NN likely to work well for most of these settings, but may be slower to train
